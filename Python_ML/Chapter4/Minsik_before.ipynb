### 3. 미분

★★★★★ 머신러닝과 미분의 중요성
- Mechine Learning의 궁극적인 목적 "최적화 문제(최소 혹은 최대가 입력 찾는 문제)"를 찾는 것, 이 과정에서 사용되는 요소가 함수의 기울기(미분)이다.

#### (1) 미분 표현

- f(w)의 w에 대한 미분율
$$  {df(w) \over dw} = {d \over dw} f(w) = f^\prime(w) $$

#### (2) 마분의 수식 변형
- 상수(스칼라)배가 곱해진 경우는 숫자는 미분 기호 왼쪽의 표시
- 미분과 무관한 부분(미분하고자 하는 변수 이외의 변수)는 문자식 왼쪽 표시(바깥으로)
- 미분하고자 하는 변수가 없는 함수의 미분 값은 0

#### (3) 중첩된 함수의 미분: 연쇄 법칙

$$ {d \over dw} f(g(w)) = {df \over dg} {dg \over dw} $$ 

#### (4) 편미분
- 의미: 다변수 함수의 특정 변수를 제외한 나머지 변수를 상수로 생각하여 미분하는 것 
$$
 f_\prime {w_0} =  {\partial f(w_0, w_1) \over \partial w_0}
$$

->$ w_0 $ 와 $ w_1 $의 두개 변수를 가진 다항식에서 $w_0$만을 변수로 주목하여 미분  
->ex) $ f(w_0, w_1) = w_0^2 + 2w_0w_1 + 3$일 경우 w_0의 편미분 ${df \over dw_0} = 2w_0 + 2w_1$ (w_1 변수는 상수 취급) 

#### (5) 편미분과 도형(기하학적 의미)
- 편미분의 기하학적인 의미 -> 편미분을 하는 변수 방향으로 기울기
- ${\partial f \over \partial w_0} w_0$는 $w_0$축가 평행하도록 $f$를 잘랐을 때의 단면, 이 때 $w_0$ 축과 평행 되는 면은 수 없이 많으며 다른 변수 $w_1$의 값이 정해지면 평행되는 수 없이 많은 면 중에서의 특정 면에 대한 기울기 값이 됨

- 각각의 변수(ex: $w_0$와 $w_1$)에 대한 기울기를 벡터로 해석하여 $f$와 $w$에 대한 경사(기울기 벡터)를 아래와 같이 일반화 할 수 있음
$$  
\triangledown _w f\
= \begin{bmatrix} {\partial f \over \partial w_0} \\ {\partial f \over \partial w_1} \end{bmatrix}
$$


- 참고 영상: https://www.youtube.com/watch?v=dcAvFFM_QQU

#### cf) 참고사항 

[1] np.meshgrid(xi, copy = True, sparse = False, indewxing = 'xy)
- xi: grid에 나타낼 1차원 배열 vector 
- copy(default = True): False가 되면 메모리 절약을 위해 오리지널 배열로 변환
- sprase(default = False): True가 되면 메모리 절약을 위해 희소 그리드 변환
- indexing('xy' or 'ij): 일반좌표(xy: x가 가로 y가 세로) vs 행력 인덱스(ij,: i가 세로 j를 가로)

[2] plt.contour(xi, yi, zi, levels: nums, linewidhts: float, colors(or cmap)
- x, y, z 모두 2차원 array(x와 y의 경우 np.meshgrid)

[3] plt.quiver(x, y, u, v, angles = 'xy', pviot = 'middle)
- x, y는 vector의 위치(1차원 vector) + np.meshgrid로 생성된 array로 
- u, v는 vector의 성분(2차원 vecotr)
- angle: vecotr 

```python
import numpy as np
import matplotlib.pyplot as plt

def f(w0, w1):             # 식 4-41을 정의 
    return w0 ** 2 + 2 * w0 * w1 + 3
def df_dw0(w0, w1):        # 식 4-41에 대한 w0에 편미분
    return 2 * w0 + 2 * w1
def df_dw1(w0, w1):        # 식 4-41에 대한 w1에 편미분
    return 2 * w0 + 0 * w1

w_range = 2 # w의 공간 크기
dw = 0.25   # step

w0 = np.arange(-w_range, w_range + dw, dw) # w0 변수에 할당할 값 분포(-2와 2 사이의 0.25 가격만큼 값을 생성)
w1 = np.arange(-w_range, w_range + dw, dw) # w1 변수에 할당할 값 분포(-2와 2 사이의 0.25 가격만큼 값을 생성)
wn = w0.shape[0]                           # w0 변수의 할당된 값 개수(17)
ww0, ww1 = np.meshgrid(w0, w1)             # np.meshgrid: 직사각형의 grid(격자) 위한 function(3차원 배열의 유용함) # wo와 w1의 격자 생성ㄴ
ff = np.zeros((len(w0), len(w1)))          # ff목적: w0와 w1 격자의 값마다의 함수값을 저장하기 위한 변수 위의 격자의 크기와 맞춰줌: w0개수 * w1개수)
dff_dw0 = np.zeros((len(w0), len(w1)))     # dff_dw0목적: w0와 w1 격자의 값마다의 w0 편미분 값들을 저장하기 위한 변수
dff_dw1 = np.zeros((len(w0), len(w1)))     # dff_dw1목적: w0와 w1 격자의 값마다의 w1 편미분 값들을 저장하기 위한 변수

for i0 in range(wn):
    for i1 in range(wn):
        # 행: W1 변수 변화 / 열: W0의 변수 변화 -> ex) 첫번째 행 array 은 w1 = -2일때 w0의 변화에 따른 함수 값들 array
        ff[i1, i0] = f(w0[i0], w1[i1])
        # 위와 상동 -> ex) 첫번재 행 array w1 = -2 일때 w0의 변화에 따른 함수 f의 w0의 경사
        dff_dw0[i1, i0] = df_dw0(w0[i0], w1[i1])
        # 위와 상동 -> ex) 첫번재 행 array w1 = -2 일때 w0의 변화에 따른 함수 f의 w1의 경사
        dff_dw1[i1, i0] = df_dw1(w0[i0], w1[i1])
        
plt.figure(figsize = (15,5))
plt.subplots_adjust(wspace = 0.5)                   # subplot 간으이 간격(2개의 subplot width = 0.3으로 벌림)

# 함수 f의 등고선을 표시(우상단 +좌하단이 높고, 좌상단과 우하단이 낮은 지형 / 등고선)
plt.subplot(1, 2, 1)
cont = plt.contour(ww0, ww1, ff, 10, colors = 'k')  #  등고선 제작 (숫자는 ff 함수 값을 표현)
cont.clabel(fmt = '%2.0f', fontsize = 8)            #  내부 등고선의 label 표시(ff함수값)
plt.xticks(range(-w_range, w_range + 1, 1))
plt.yticks(range(-w_range, w_range + 1, 1))
plt.xlim(-w_range-0.5, w_range + 0.5)
plt.ylim(-w_range-0.5, w_range + 0.5)
plt.xlabel('$w_0$', fontsize = 14)
plt.ylabel('$w_1$', fontsize = 14)
plt.title('Contour plot according to variables w0 and w1')

# 함수 f의 경사를 표시(위의 등고선과 마찬가지로 경사가 높은 쪽으로 향하고 있음)
plt.subplot(1, 2, 2)
plt.quiver(ww0, ww1, dff_dw0, dff_dw1)              # plot의 x와 y는 각각의 변수 w0와 변수 w1의 값들 / 속성: w0와 w1의 편미분 값들 
plt.xlabel('$w_0$', fontsize = 14)
plt.ylabel('$w_1$', fontsize = 14)
plt.xticks(range(-w_range, w_range + 1, 1))
plt.yticks(range(-w_range, w_range + 1, 1))
plt.xlim(-w_range-0.5, w_range + 0.5)
plt.ylim(-w_range-0.5, w_range + 0.5)
plt.show()
```

#### (6) 다변수의 중첩 함수 미분 
- 앞선 일변수 함수간의 중첩 함수미분과 동일하게 "연쇄법칙"을 활용하여 문제를 해결함

$$  {\partial f(g_0(w_0, w_1),g_1(w_0, w_1)) \over \partial w_0} = 
{df \over dg_0} {dg_0 \over dw_0} + {df \over dg_1} {dg_1 \over dw_0} $$ 

-> $ w_0 $에 대해 편미분을 진행하면 $w_0$가 있는 각각의 함수를 연쇄법칙을 적용하여 값을 구한 뒤 더해줌

  
- 2개 이상 함수가 중첩될 경우의 연쇄법칙은 다음과 같이 일반화 가능 
$$ 
{df \over dw_0}
= {df \over dg_0} {dg_0 \over dw_0} + {df \over dg_1} {dg_1 \over dw_0} \cdots + {df \over dg_M} {dg_M \over dw_0} 
= \sum_{m=0}^M {{df \over dg_m} {dg_m \over dw_0}}
$$ 

#### (7) 합과 미분의 교환 가능: 미분과 합의 기호는 순서를 바꿀 수 있음(★)

- 미분 기호를 합의 기호 앞으로 이동할 수 있음 (합 계산 후 미분 -> 각 항별로 미분 후 더함)
$$ 
{\partial \over \partial w} \sum_{n} {f_n(w)}
= \sum_{n} {\partial \over \partial w} {f_n(w)}
$$

#### 4. 행렬의 곱셈

[0] 행렬의 곱셈의 전제조건: (행렬 A의 열의 수 = 행렬 B의 행의 수) 

[1] 수식

행렬 A가 $ L \times M $ 행렬이고 행렬 B가 $M \times N $일 때
$$ \begin{bmatrix} AB_{i,j} \end{bmatrix} = \sum_{m=0}^M a_{i, m} a_{m, n}  $$

[2] 성질 
- 일반적으로 AB = BA 계산 불가함   

    Ex) A가 $ 2 \times 3 $ 행렬이고 B = $ 3 \times 4 $ 행렬이면   
    $\Rrightarrow$ AB 가능(A의 열의 수 = B의 행의 수)하지만 BA는 불가(A의 행의 수 $\ne$ B의 열의 수) 


- 단 행렬 A와 B가 정방 행렬(=정사각형 행렬, 행과 열의 개수가 같은 행렬)일 경우 AB와 BA 모두 계산할 수 있지만 AB = BA 성립되지 않음  
(실수의 곱셈과 다른 부분 / 실수의 곱셈에서는 a * b = b * a)

[3] 참고사항
- 보통의 곱셈 기호(*)를 사용하는 것은 위에서 언급한 행렬의 곱셈이 아님. 단지 행렬의 각 요소마다 곱하는 연산

[3] 로그 활용

(1) 로그는 곱셈을 덧셈으로 변환 ->로그 우도 발전

$ \Rrightarrow \log \prod_{n=1}^N f(n)  \\
= \log(f(1) \times f(2) \cdots \times f(N-1) \times f(N)) \\
= \log f(1) + \log f(2) \cdots + \log f(N-1) + \log f(N)  \\
= \sum_{n=1}^N \log f(n) $  
  
  
(2) 로그 최소화: $ argmin f(x) = argmin \log f(x)$
- 함수 $f(x)$를 최소(최대)화하는 $ x = \dot{x} $ 이면, 로그를 취한 $\log f(x)$도 $ x = \dot{x} $일 때 최소(최대)화된다.
- 매우 큰 수 혹은 작은 수에 대해서 log를 활용해 최적화할 수 있음

```python
x = np.linspace(-4, 4, 100)
y = (x - 1) ** 2 + 2
logy = np.log(y)

plt.figure(figsize = (5,5))
plt.plot(x, y, 'black', linewidth = 3)             # 정의된 함수 시각화
plt.plot(x, logy, 'cornflowerblue', linewidth = 3) # 정의된 함수의 log 시각화
plt.xticks(range(-4, 4 ,1))
plt.yticks(range(-4, 9 ,1))
plt.xlim(-4, 4)
plt.ylim(-4, 9)
plt.grid(True)

# 그래프 설명 표시(수식)
plt.title('$ The minimum value of the equation: f(x) = (x-1)^2 +1$ ')
plt.text(1.0, 5.0, '$ f(x) $', fontsize = 15)
plt.text(-1.0, 0.0, '$ \log f(x) $', fontsize = 15)
plt.show()
```

#### (5) softmax(소프트맥스) 함수

- N개의 수의 대소 관계를 유지하면서 각각 확률(0과 1사이의 숫자이며 N개의 확률을 더할 때 1이되어야함)로 변환하는 함수

[1] 함수
$$ y_{i} = exp(x_{i}) \over  \sum_{j=0}^{K-1} exp(x_{j}) $$   

[2] 미분 
$$ {dy_{j} \over dx_{i}}  = y_{j}(I_{ij}-y_{i})$$  

[3] softmax 함수와 sigmoid 함수의 관계: sigmoid 함수를 다변수로 확장한 것 -> softmax 함수 

cf) 참고사항
- ax.plot_surface(x값, y값, z값, row step size, column step size, alpha. color, edgecolor)
-> alpha: 투명도 / edgecolor: 선의 색 / step size: plot의 step 크기 

```python
def softmax(x0, x1, x2):  # 3개의 입력에 대한 softmax 변환
    u = np.exp(x0) + np.exp(x1) + np.exp(x2)               # 확률로 변하기 위한 각 xi의 exp 합계 구하기
    return np.exp(x0) / u, np.exp(x1) / u, np.exp(x2) / u # u로 나눔으로써 각각의 x에 대한 확률 구하긴

y = softmax(2, 1, -1)
print(np.round(y, 2)) # 소수점 2자리까지 확인
print(np.sum(y))      # 합계가 1인 확률인지 확인
```

- (왼쪽 그래프)    $x_{2} = 1$로 고정하고 $x_{0}과 x_{1}$의 변화에 따른 $y_{0}$ 변화 plot  
  (ex- 우상단: $x_{1}$ 값이 작을수록 z값이 높아짐)
    
- (오른쪽 그래프)  $x_{2} = 1$로 고정하고 $x_{0}과 x_{1}$의 변화에 따른 $y_{0}$ 변화 plot  
  (ex- 좌상단: $x_{0}$ 값이 작을수록 z값이 높아짐)

```python
from mpl_toolkits.mplot3d import Axes3D   # 3차원 시각화 Module
 
xn = 20
x0 = np.linspace(-4, 4, xn)
x1 = np.linspace(-4, 4, xn)

y = np.zeros((xn, xn, 3))   

for i0 in range(xn):       # x0을 기준으로 x0의 값들, x1의 값들, 1에 대한 softmax를 구한 후 y 저장(예: (-4, -4, 1))
    for i1 in range(xn):
        y[i1, i0, :] = softmax(x0[i0], x1[i1], 1)  
        
xx0, xx1 = np.meshgrid(x0, x1)                # x0와 x1의 값들만큼 격자 생성

plt.figure(figsize = (10, 5))
for i in range(2):                           # range(2) 로 한 이유: y는 3개(w0, w1, 1에 대한 변화량) 中 w0, w1에 대한 변화량 시각화
    ax = plt.subplot(1, 2, i + 1, projection = '3d')  # projection = '3d'은 공간을 3차원 형태로 표현
    ax.plot_surface(xx0, xx1, y[:, :, i], rstride = 1, cstride = 1, alpha = 0.3, color = 'blue', edgecolor = 'black')
    ax.set_xlabel('$x_0$',fontsize = 14)
    ax.set_ylabel('$x_1$',fontsize = 14)
    ax.view_init(40,-125)
```

#### (6) Gaussain(가우스) 함수 = 정규분포 함수

함수
$$ y = a * exp({-(x-u)^2 \over \sigma ^2}) $$   

- $a$: 높이
- $u$: 중심(평균) - 종의 중심위치를 결정
- $\sigma$: 확산(표준편차) - 종의 너비를 결정

- 정규분포는  평균과 표준편차에 대해 모양이 결정되고, 이때의 분포를 $N(u, \sigma^2)$로 표기
- 평균이 0이고 표준편차가 1인 정규분포를 표준 정규 분포

-> 적분해서 1이 되기 위한 필요충분조건: $ a = {1 \over (2\pi\sigma^2)^{1 \over 2}}$  
-> why? https://ride-or-die.info/gaussian-distribution-integral/

```python
def gauss(mu, sigma, a):
    return a * np.exp(-(x - mu) ** 2 / sigma ** 2)
    
x = np.linspace(-4, 4, 100)
plt.figure(figsize = (4, 4))
plt.plot(x, gauss(0, 1, 1), 'black', linewidth = 3)
plt.plot(x, gauss(2, 3, 0.5), 'gray', linewidth = 3)
plt.xlim(-4, 4)
plt.ylim(-.5, 1.5)
plt.grid(True)
plt.text(-2.0, 1.25, '$ exp(-x^2) $', fontsize = 15)
plt.show()
```

#### (7) 2차원 가우스 함수 

함수
$$ y = a * exp({1 \over 2}(x-u)^T (\sum)^-1(x-u)) $$   

- x = $\begin{bmatrix} x_0, x_1 \end{bmatrix}$ (입력 vector / 차원마다의 입력값)
- u = $\begin{bmatrix} u_0, u_1 \end{bmatrix}$ (평균 vector / 차원마다의 평균값, 분포 중심)
- $(\sum)$ =  $\begin{bmatrix} \sigma_0, \sigma_2 \\ \sigma_2, \sigma_1 \end{bmatrix}$ (공분산 행렬, 확률변수의 상관정도)  
$\sigma_0$과 $\sigma_1$은 각각의 $x_0$와 $x_1$ 방향의 분포 퍼짐(각각의 변수들의 분산의 개념)  
$\sigma_2$는 $x_0$와 $x_1$ 상관관계의 대응(변수들끼리의 선형관계를 측도(양의 상쪽[오른쪽으로 왜곡], 음의 상관[왼쪽으로 왜곡], 0(서로 독립)

-> 적분해서 1이 되기 위한 필요충분조건: $ a = {1 \over (2\pi)^{D \over 2}}{1 \over\begin{vmatrix} (\sum)^{1 \over 2} \end{vmatrix}}$

- 참고자료: https://dhpark1212.tistory.com/entry/%EB%8B%A4%EB%B3%80%EB%9F%89-%EA%B0%80%EC%9A%B0%EC%8B%9C%EC%95%88-%EB%B6%84%ED%8F%ACMultivariate-Gaussian-Distribution

```python
import numpy as np
import matplotlib.pyplot as plot
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline

def gauss(x, mu, sigma): 
    N, D = x.shape                              
    c1 = 1 / (2 * np.pi) ** (D / 2)           # 알파부분1: 1/ (2 * pi)에 '데이터 차원 / 2'만큼 제곱.
    c2 = 1 / (np.linalg.det(sigma) ** (1 / 2))# 알파부분2: 공분산행렬의 행렬식 역수 값에 '1/2'만큼 제곱
    inv_sigma = np.linalg.inv(sigma)          # 공분산 행렬의 역행렬 구함
    c3 = x - mu                               # x-u 벡터를 구함
    c4 = np.dot(c3, inv_sigma)                # 각 변수의 2차식 형태로 만들기 위한 전치
    c5 = np.zeros(N)                          # 알파부분을 합치기 위한 정의
    for d in range(D):                     
        c5 = c5 + c4[:, d] * c3[:, d]         # 알파 정의 
    
    p = c1 * c2 * np.exp(-c5 / 2)
    return p
```

```python
x = np.array([[1, 2], [2, 1], [3, 4]])   # 입력 vector
mu = np.array([1, 2])                    # 각 변수의 평균 vector
sigma = np.array([[1, 0], [0, 1]])       # 공분산 행렬(각각의 변수의 분산 = 1, 각 변수끼리의 상곤계수는 0 독립인 상황)
print(gauss(x, mu, sigma))              
```

```python
X_range0=[-3,3]
X_range1=[-3,3]

# 등고선 표시 --------------------------------
def show_contour_gauss(mu, sig):
    xn = 40 # 등고선 표시 해상도
    x0 = np.linspace(X_range0[0], X_range0[1], xn)       # x0 array 값 생성
    x1 = np.linspace(X_range1[0], X_range1[1], xn)       # x1 array 값 생성
    xx0, xx1 = np.meshgrid(x0, x1)                       # 격자 생성(40 by 40)
    x = np.c_[np.reshape(xx0, xn * xn, 'F'), np.reshape(xx1, xn * xn, 'F')] # 40 ny 40 shape을 1600의 1차원 배열로 변경 (mu 차원 맞추기 위함)
    f = gauss(x, mu, sig)                                # gauss 적용
    f = f.reshape(xn, xn)                                # 원 형태로 변환
    f = f.T                                              # 전치를 해줘야 x축방향이 x0 y축 방향이 x1dl ehldjtjt qhs rmflaeofh rksmd
    cont = plt.contour(xx0, xx1, f, 15, colors='k')      # 등고선 제작
    plt.grid(True)

# 3D 표시 ----------------------------------
def show3d_gauss(ax, mu, sig):
    xn = 40 # 등고선 표시 해상도
    x0 = np.linspace(X_range0[0], X_range0[1], xn)        
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)
    x = np.c_[np.reshape(xx0, xn * xn, 'F'), np.reshape(xx1, xn * xn, 'F')]
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    f = f.T                                              # 위와 상동ㅉ
    ax.plot_surface(xx0, xx1, f,
                    rstride=2, cstride=2, alpha=0.3,
                    color='blue', edgecolor='black')

# 메인 -----------------------------------
mu = np.array([1, 0.5])
sigma = np.array([[2, 1], [1, 1]])

# 등고선 표현
Fig = plt.figure(1, figsize=(7, 3))
Fig.add_subplot(1, 2, 1)
show_contour_gauss(mu, sigma)
plt.xlim(X_range0)
plt.ylim(X_range1)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)

# 3D 표시
Ax = Fig.add_subplot(1, 2, 2, projection='3d')
show3d_gauss(Ax, mu, sigma)
Ax.set_zticks([0.05, 0.10])
Ax.set_xlabel('$x_0$', fontsize=14)
Ax.set_ylabel('$x_1$', fontsize=14)
Ax.view_init(40, -100)
plt.show()
```

```python
# 다른 예시 제작(그림 4-36)
X_range0=[-3,3]
X_range1=[-3,3]

mu = np.array([0, 0])
sigma = np.array([[1, 0], [0, 1]])

# 등고선 표현
Fig = plt.figure(1, figsize=(7, 3))
Fig.add_subplot(1, 2, 1)
show_contour_gauss(mu, sigma)
plt.xlim(X_range0)
plt.ylim(X_range1)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)

# 3D 표시
Ax = Fig.add_subplot(1, 2, 2, projection='3d')
show3d_gauss(Ax, mu, sigma)
Ax.set_zticks([0.05, 0.10])
Ax.set_xlabel('$x_0$', fontsize=14)
Ax.set_ylabel('$x_1$', fontsize=14)
Ax.view_init(40, -100)
plt.show()
```
